```python
import torch

# ---- 1. Load checkpoint ----
ckpt_path = "../param/temp/epoch3_20251017-020145.pt"   # replace with your file
ckpt = torch.load(ckpt_path, map_location="cpu")

# Sometimes ckpt is a dict with 'state_dict' key (from trainer wrappers)
if "state_dict" in ckpt:
    state_dict = ckpt["state_dict"]
else:
    state_dict = ckpt

# ---- 2. Print layer names ----
print("Layers in checkpoint:")
for name in state_dict.keys():
    print(name)
```

```python

# ---- 3. Print weights of a specific layer ----
layer_name = "dist_head.beta"   # replace with actual name from printout
if layer_name in state_dict:
    print(f"\nWeights for {layer_name}:")
    print(state_dict[layer_name])
else:
    print(f"Layer {layer_name} not found.")
```
