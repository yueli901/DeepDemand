```python
from pathlib import Path
import zipfile

# Root folder containing the "subgraphs" directory
# If your notebook is already in the same root, this can stay "."
PROJECT_ROOT = Path("..")

SUBGRAPHS_DIR = PROJECT_ROOT / "subgraphs"
ZIP_OUT = PROJECT_ROOT / "subgraphs_meta_only.zip"

SUBGRAPHS_DIR, ZIP_OUT
```

```python
if not SUBGRAPHS_DIR.is_dir():
    raise RuntimeError(f"{SUBGRAPHS_DIR} is not a directory.")

count_added = 0

with zipfile.ZipFile(ZIP_OUT, mode="w", compression=zipfile.ZIP_DEFLATED) as zf:
    # iterate over all immediate subdirectories under subgraphs/
    for edge_dir in SUBGRAPHS_DIR.iterdir():
        if not edge_dir.is_dir():
            continue

        meta_path = edge_dir / "meta.json"
        if not meta_path.is_file():
            # skip if this edge directory has no meta.json
            continue

        # keep folder hierarchy starting from project root, e.g. subgraphs/123_456_0/meta.json
        arcname = meta_path.relative_to(PROJECT_ROOT)
        zf.write(meta_path, arcname=arcname)
        count_added += 1

print(f"Done. Added meta.json from {count_added} edge directories into {ZIP_OUT}")
```
