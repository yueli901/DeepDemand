```python
import os

def get_unique_extensions(root_dir="."):
    extensions = set()

    for root, dirs, files in os.walk(root_dir):
        for file in files:
            _, ext = os.path.splitext(file)
            if ext:  # ignore files without extension
                extensions.add(ext.lower().lstrip("."))

    return sorted(extensions)


if __name__ == "__main__":
    unique_exts = get_unique_extensions(".")
    print("Unique file extensions found:")
    for ext in unique_exts:
        print(ext)
```
